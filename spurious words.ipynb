```python
#筛选出含有二声和三声的词语
import pandas as pd
import numpy as np
df = pd.read_csv('chineselexicaldatabase2.1.csv')
filter = df['Pinyin'].str.contains('2|3',regex=True)
df = df[filter]
newdf = df[["Word", "Pinyin", 'Length', 'Frequency']]
newdf['Pinyin'].replace('\d', '', regex=True, inplace = True)
newdf['Pinyin2'] = df['Pinyin']
newdf = newdf[["Word", "Pinyin",'Pinyin2', 'Length', 'Frequency']]
```

```python
# 把newdf二声都变成三声
newdf['Pinyin3'] = newdf['Pinyin2'].replace('2', '3', regex=True)

# 第一步：把双音节词和三、四音节词分开
spurdf = newdf[newdf['Length'] == 2]
# spurdf.shape[0] 得到22364个
spurdf_1 = newdf[newdf['Length'] == (3 or 4)]
# spurdf_1.shape[0] 得到5557个
```

```python
# 第二步：去掉spurdf中没有变化的words
# 去掉没有变化的words
spurdf_2 = spurdf[newdf['Pinyin2'] != newdf['Pinyin3']]
```

```python
# 第三步：把这些words放进三、四音节词匹配,得到新的dataframe

# 先新建一个empty dataframe spurdf_3
spurdf_3 = pd.DataFrame(columns = spurdf_1.columns)
# 然后匹配，得到新的dataframe和spurdf_2的index，放进spurdf_3
for index, row in spurdf_2.iterrows():
    ex = row['Pinyin3']
    filter = spurdf_1['Pinyin2'].str.contains('(^|\d)' +str(ex), regex=True)
    data = spurdf_1[filter]
    if not data.empty:
        spurdf_3  = pd.concat([spurdf_3, spurdf_2.loc[[index]], data])
```

```python
spurdf_4 = spurdf_3[spurdf_3['Length'] == 3]
spurdf_4.sort_index(inplace=True)
spurdf_4
```

```python
mask = spurdf_4.duplicated(['Pinyin2'])
spurdf_5 = spurdf_4[~mask]
spurdf_5.shape[0]
```
