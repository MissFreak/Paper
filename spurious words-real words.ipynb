```python
#筛选出含有二声和三声的词语
import pandas as pd
import numpy as np
df = pd.read_csv('chineselexicaldatabase2.1.csv')
df = df[["Word", "Pinyin", 'Length', 'Frequency']]
```

```python
# 第一步：把双音节词和三、四音节词分开
spurdf = df
spurdf_1 = spurdf[spurdf['Length'] == 2]
spurdf_2 = spurdf[spurdf['Length'] == 3]
```

```python
# 第三步：把这些words放进三、四音节词匹配,得到新的dataframe

# 先新建一个empty dataframe spurdf_3
spurdf_3 = pd.DataFrame(columns = spurdf_1.columns)
# 然后匹配，得到新的dataframe和spurdf_2的index，放进spurdf_3
for index, row in spurdf_1.iterrows():
    ex = row['Pinyin']
    filter = spurdf_2['Pinyin'].str.contains('(^|\d)' +str(ex), regex=True)
    data = spurdf_2[filter]
    if not data.empty:
        spurdf_3  = pd.concat([spurdf_3, spurdf_1.loc[[index]], data])
```
