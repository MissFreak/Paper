```python
#筛选出含有二声和三声的词语
import pandas as pd
import numpy as np
df = pd.read_csv('chineselexicaldatabase2.1.csv')
df = df[["Word", "Pinyin", 'Length',  'Frequency']]
df.sort_values(by=['Pinyin', 'Length','Frequency'], ascending=[True, True,False], inplace=True)
df
```

```python
df[df['Pinyin'] == 'ao2']
```

```python
# 先得到所有音节重复的单词个数
mask = df.duplicated(['Pinyin'], keep = False)
newdf = df[mask]
# 计算real homophony的个数
newdf.shape[0]
```

```python
newdf
```

```python
newdf.to_csv ('real_homophone.csv', header=True)
```

```python
newdf['Length'].value_counts()
```

```python
newdf['Pinyin'].replace('\d', '', regex=True, inplace = True)
```

```python
newdf['Pinyin'].value_counts()
```

```python
# 先得到所有音节重复的单词个数
mask2 = newdf.duplicated(['Pinyin'])
newdf2 = newdf[~mask2]
# 计算real homophony的个数
newdf2.shape[0]
```

```python
newdf2
```

```python
newdf2.to_csv ('real-homophone-target.csv', header=True)
```

```python
newdf3 = pd.concat([newdf2, newdf]).drop_duplicates(keep=False)
```

```python
newdf3.to_csv ('real-homophone-competitors.csv', header=True)
```
